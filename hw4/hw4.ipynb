```python
# import class
from drv import DRV  
```

```python
# formula N = R* * fp * ne * f1 * fi * fc * L

# R*
R = DRV(type='normal', mean=2.25, stdev=0.43, bins=30)
print("R:", R.ev())
```

```python
# fp
fp = DRV(dist={.85: 0.25, 0.90: 0.25, 0.95: 0.50}, type='discrete')
print("fp:", fp.ev())
```

```python
# ne
ne = DRV(type='uniform', min=1, max=5, bins=5)
print("ne:", ne.ev())
```

```python
# f1
f1 = DRV(type='uniform', min=0, max=1, bins=5)
print("f1:", f1.ev())
```

```python
# fi 
fi = DRV(type='uniform', min=0, max=1, bins=5)
print("fi:", fi.ev())
```

```python
# fc 
fc = DRV(type='uniform', min=0, max=1, bins=5)
print("fc:", fc.ev())
```

```python
# L
L = DRV(type='normal', mean=10000, stdev=2000, bins=30)
print("L:", L.ev())
```

```python
# find N
N = R * fp * ne * f1 * fi * fc * L
print(N)
```

```python
# find N expected value & standard deviation 
N_ev = N.ev()
print("N ev:", N_ev)
print("N. std:", N.stdev())
```

```python
N.plot(cumulative=False, log_scale=True, trials=0, bins=100)
```
